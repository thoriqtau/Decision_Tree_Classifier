```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/AML_PAD/pilkada-1000-uas.csv')
df.head()
```

```python
df = df.drop(['Nama Daerah','Persentase PARPOL Pendukung'], axis=1)
df.head()
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# membuat encoding untuk kolom hasil pilkada
hasil_pilkada = LabelEncoder().fit_transform(df['Hasil Pilkada'])
df['Hasil Pilkada'] = hasil_pilkada

# membuat encoding untuk kolom jalur pencalonan
jalur_pencalonan = LabelEncoder().fit_transform(df['Jalur Pencalonan'])
df['Jalur Pencalonan'] = jalur_pencalonan

# membuat encoding untuk kolom nama kandidat (1)
nama_kandidat_1 = LabelEncoder().fit_transform(df['Nama Kandidat (1)'])
df['Nama Kandidat (1)'] = nama_kandidat_1

# membuat encoding untuk kolom nama kandidat (2)
nama_kandidat_2 = LabelEncoder().fit_transform(df['Nama Kandidat (2)'])
df['Nama Kandidat (2)'] = nama_kandidat_2

# membuat encoding untuk kolom nama gender (1)
nama_gender_1 = LabelEncoder().fit_transform(df['Gender (1)'])
df['Gender (1)'] = nama_gender_1

# membuat encoding untuk kolom nama gender (2)
nama_gender_2 = LabelEncoder().fit_transform(df['Gender (2)'])
df['Gender (2)'] = nama_gender_2

# membuat encoding untuk kolom latar belakang (1)
latar_belakang_1 = LabelEncoder().fit_transform(df['Latar Belakang Profesi (1)'])
df['Latar Belakang Profesi (1)'] = latar_belakang_1

# membuat encoding untuk kolom latar belakang (1)
latar_belakang_2 = LabelEncoder().fit_transform(df['Latar Belakang Profesi (2)'])
df['Latar Belakang Profesi (2)'] = latar_belakang_2

df.head()
```

# Model

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Hasil Pilkada', axis=1)
y = df['Hasil Pilkada']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model_dtr = DecisionTreeClassifier()

parameters = {
    'random_state': [42],
    'max_depth': [1, 2, 3, 4, 5],
    'criterion': ['gini', 'entropy'],
    'min_samples_split': [2,3,4,5],
    'min_samples_leaf': [1,2,3,4,5],
    'max_features': ['sqrt'],
}
model = GridSearchCV(model_dtr, parameters, cv=3, n_jobs=-1, verbose=1)
model.fit(x_train, y_train)
print(model.best_params_)
print(model.score(x_train,y_train), model.score(x_test,y_test))
```

```python
pd.DataFrame(model.cv_results_).sort_values('rank_test_score').head()
```
